```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_df = pd.read_csv("animal longevity.csv")
data_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

data_df.drop(data_df.head(27).index,inplace=True)
data_df.drop(data_df.tail(1).index,inplace=True)

data_df['Temperature (K)'] = ((np.abs(data_df['Temperature (K)'])-273.15) * 1.8) + 32
data_df = data_df.rename(columns = {'Temperature (K)':'Temperature', 'Common name':'Name', 'Birth weight (g)':'Birth weight', 
                                    'Adult weight (g)':'Adult weight', 'Maximum longevity (yrs)':'Longevity',
                                    'Specimen origin':'Origin', 'Body mass (g)':'Body mass', 'Metabolic rate (W)':'Metabolic rate',
                                   'Male maturity (days)':'Male maturity', 'Female maturity (days)':'Female maturity'})

data_df = data_df.drop(['References',
                        'Litter/Clutch size',
                        'Litters/Clutches per year', 
                        'Inter-litter/Interbirth interval',
                        'Source',
                        'IMR (per yr)',
                        'MRDT (yrs)', 
                        'Sample size', 
                        'Growth rate (1/days)', 
                        'Data quality', 
                        'Weaning weight (g)', 
                        'Gestation/Incubation (days)', 
                        'Weaning (days)', 'Family', 'Name', 'Genus', 'Species'], axis = 1)

for column in ['Kingdom', 'Phylum', 'Class']:
    dummies = pd.get_dummies(data_df[column], prefix=column)
    data_df = pd.concat([data_df, dummies], axis=1)
    data_df = data_df.drop(column, axis=1).reset_index(drop=True)

class_labels = LabelEncoder()
data_df['Origin'] = class_labels.fit_transform(data_df['Origin'].values)

data_df['Temperature'].fillna(data_df['Temperature'].mean(), inplace=True)
data_df['Adult weight'].fillna(data_df['Adult weight'].mean(), inplace=True)
data_df['Birth weight'].fillna(data_df['Birth weight'].mean(), inplace=True)

data_df['Body mass'] = data_df['Body mass'].interpolate()
data_df['Longevity'] = data_df['Longevity'].interpolate()
data_df['Female maturity'] = data_df['Female maturity'].interpolate()
data_df['Male maturity'] = data_df['Male maturity'].interpolate()
data_df['Metabolic rate'] = data_df['Metabolic rate'].interpolate()

data_df['Birth weight'] = np.abs(data_df['Birth weight']) / 1000
data_df['Adult weight'] = np.abs(data_df['Adult weight']) / 1000
data_df['Male maturity'] = np.abs(data_df['Male maturity']) / 365
data_df['Female maturity'] = np.abs(data_df['Female maturity']) / 365



data_df.head()
```
